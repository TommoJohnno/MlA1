```python
HospitalDataTrain = pd.read_csv('./train_data.csv')
HospitalDataTrain.head()
```

```python
plt.figure(figsize=(20,20))
for i, col in enumerate(HospitalDataTrain.columns[2:]):
    plt.subplot(4,5,i+1)
    plt.hist(HospitalDataTrain[col], alpha=0.3, color='b', density=True)
    plt.title(col)
    plt.xticks(rotation='vertical')
```

```python
from sklearn.preprocessing import OneHotEncoder


columns_to_replace = ['Gender', 'Race', 'TypeOfAdmission', 'PaymentTypology', 'EmergencyDepartmentIndicator']
for column in columns_to_replace:
    OneHotEncoder_column = OneHotEncoder(handle_unknown='ignore')
    OneHotEncoder_column.fit(HospitalDataTrain[[column]])

    onehot_ = OneHotEncoder_column.transform(HospitalDataTrain[[column]]).toarray()
    for i in range(len(OneHotEncoder_column.categories_[0])):
        HospitalDataTrain[column + '_' + str(OneHotEncoder_column.categories_[0][i])] = onehot_[:,i]
    
    HospitalDataTrain.drop(columns = [column], inplace=True)
#     print(onehot_.shape, OneHotEncoder.categories_)
# HospitalDataTrain.drop(columns = columns_to_replace, axis=1)
HospitalDataTrain.head()
```

```python
HospitalDataTrain_X = HospitalDataTrain.drop(columns =["ID", "HealthServiceArea", "LengthOfStay"])
HospitalDataTrain_Y = HospitalDataTrain['LengthOfStay']
```

```python
from sklearn.model_selection import train_test_split

with pd.option_context('mode.chained_assignment', None):
    train_X_train, train_X_test, test_y_train, test_y_test = train_test_split(HospitalDataTrain_X,HospitalDataTrain_Y, test_size=0.2, shuffle=True)
```
